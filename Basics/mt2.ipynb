```python
import os, sys, time, threading, multiprocessing, random as r

# Function runs for random seconds
def task(cmd):

    waitingTime = r.randint(5,10)

    print("  Task started")
    print("   Task running for %d sec...." % (waitingTime))
    time.sleep(waitingTime)

    print("  Task ends")
    return
```

```python
# Main Program
startTime = time.time()

cmd = 'Sample task'

print("Program Started")
print(" Thread starts")

t = threading.Thread(target=task , args=(cmd,))
t.start()

print(" Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

```python
# Main Program
startTime = time.time()

activeThreads = threading.active_count()
print("Active Threads = ", activeThreads)

cmd = 'Sample task'

print("Program Started")
print(" Thread starts")

t = threading.Thread(target=task , args=(cmd,))
t.start()
time.sleep(1)

# Waiting to finish the Threads
while True:
    if threading.active_count() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.active_count() - activeThreads))
        time.sleep(1)
  
print(" Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

```python
# Main Program - Run 16 threads in one go

startTime = time.time()

numberOfThreads = 16

# Main Program - Run Multiple Threads
print("Program Started....")

for i in range(numberOfThreads):
    cmd = str(i+1)
    print(" Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()

time.sleep(1)

# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.activeCount() - activeThreads))
        time.sleep(1)
  
print(" All Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

```python
# Main Program - Run Multiple Threads = number of cores

startTime = time.time()

numberOfThreads = 16

activeThreads = threading.active_count()
print("Active Threads = ", activeThreads)

numberOfCores=multiprocessing.cpu_count()
print("Number Of Cores = ", numberOfCores)

# Main Program - Run Multiple Threads
print("Program Started....")

for i in range(numberOfThreads):
    cmd = str(i+1)
    print(" Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()
    # control the number of active threads based on the number of available CPU cores
    while True:
      if threading.active_count() - activeThreads + 1 <= numberOfCores:
        break
      time.sleep(1)

time.sleep(1)

# Waiting to finish all Threads
while True:
    if threading.active_count() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.active_count() - activeThreads))
        time.sleep(1)
  
print(" All Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```
